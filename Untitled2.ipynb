```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from math import pi

cp = 4186.0
D_inner = 12.5e-3
L = 1.5
A = pi * D_inner * L

data = [
    {"run_id": 1, "mode": "parallel", "Th_in_C": 49.5, "Th_out_C": 45.3, "Tc_in_C": 24.0, "Tc_out_C": 30.0, "mh_kg_s": 2.0, "mc_kg_s": 2.0},
    {"run_id": 2, "mode": "parallel", "Th_in_C": 48.2, "Th_out_C": 45.0, "Tc_in_C": 25.7, "Tc_out_C": 30.0, "mh_kg_s": 2.5, "mc_kg_s": 2.5},
    {"run_id": 3, "mode": "counter", "Th_in_C": 50.0, "Th_out_C": 45.6, "Tc_in_C": 31.0, "Tc_out_C": 28.2, "mh_kg_s": 2.0, "mc_kg_s": 2.0},
    {"run_id": 4, "mode": "counter", "Th_in_C": 53.5, "Th_out_C": 48.0, "Tc_in_C": 32.0, "Tc_out_C": 29.2, "mh_kg_s": 2.5, "mc_kg_s": 2.5},
]
df = pd.DataFrame(data)

def lmtd(dT1, dT2):
    if dT1 <= 0 or dT2 <= 0:
        return np.nan
    if abs(dT1 - dT2) < 1e-9:
        return dT1
    return (dT1 - dT2) / np.log(dT1/dT2)

def calc_LMTD(mode, Th_in, Th_out, Tc_in, Tc_out):
    mode = str(mode).lower()
    if mode.startswith("par"):
        dT1, dT2 = Th_in - Tc_in, Th_out - Tc_out
    else:
        dT1, dT2 = Th_in - Tc_out, Th_out - Tc_in
    return lmtd(dT1, dT2)

def Q_hot(mh, Th_in, Th_out): return mh * cp * (Th_in - Th_out)
def Q_cold(mc, Tc_in, Tc_out): return mc * cp * (Tc_out - Tc_in)

def NTU_effectiveness(UA, mh, mc, mode):
    mode = str(mode).lower()
    Ch, Cc = mh*cp, mc*cp
    Cmin, Cmax = min(Ch, Cc), max(Ch, Cc)
    if Cmin <= 0 or np.isnan(UA):
        return np.nan, np.nan
    Cr, NTU = Cmin/Cmax, UA/Cmin
    if mode.startswith("par"):
        eps = (1 - np.exp(-NTU*(1+Cr))) / (1+Cr)
    else:
        if abs(Cr - 1) < 1e-8:
            eps = NTU/(1+NTU)
        else:
            expterm = np.exp(-NTU*(1-Cr))
            eps = (1-expterm)/(1-Cr*expterm)
    return NTU, eps

results = []
for _, r in df.iterrows():
    mode = str(r['mode']).lower()
    Qh = Q_hot(r['mh_kg_s'], r['Th_in_C'], r['Th_out_C'])
    Qc = Q_cold(r['mc_kg_s'], r['Tc_in_C'], r['Tc_out_C'])
    err = 100*(Qh-Qc)/Qh if Qh != 0 else np.nan
    LMTD = calc_LMTD(mode, r['Th_in_C'], r['Th_out_C'], r['Tc_in_C'], r['Tc_out_C'])
    UA = Qh/(A*LMTD) if (A>0 and LMTD>0) else np.nan
    NTU, eff = NTU_effectiveness(UA, r['mh_kg_s'], r['mc_kg_s'], mode)
    results.append({**r.to_dict(), "Q_hot_W": Qh, "Q_cold_W": Qc, "Error_%": err, "LMTD": LMTD, "UA": UA, "NTU": NTU, "Effectiveness": eff})

res_df = pd.DataFrame(results)
display(res_df)

plt.figure()
plt.scatter(res_df['NTU'], res_df['Effectiveness'], c=res_df['run_id'], cmap='viridis')
plt.xlabel("NTU"); plt.ylabel("Effectiveness ()")
plt.title("Effectiveness vs NTU"); plt.colorbar(label="Run ID"); plt.grid(True); plt.show()

plt.figure()
plt.scatter(res_df['mh_kg_s'], res_df['UA'], label="Hot flow")
plt.scatter(res_df['mc_kg_s'], res_df['UA'], label="Cold flow")
plt.xlabel("Mass flow rate (kg/s)"); plt.ylabel("UA (W/m-K)")
plt.title("UA vs Mass flow rates"); plt.legend(); plt.grid(True); plt.show()

plt.figure()
plt.bar(res_df['run_id'].astype(str), res_df['Error_%'])
plt.xlabel("Run ID"); plt.ylabel("Energy balance error (%)")
plt.title("Energy balance error"); plt.grid(True); plt.show()

```
